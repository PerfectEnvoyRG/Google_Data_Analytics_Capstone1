#### Convert format of started_at and ended_at to %Y-%m-%d %H:%M:%S

```python
merged_df["started_at"] = pd.to_datetime(merged_df["started_at"], format='%Y-%m-%d %H:%M:%S')
merged_df["ended_at"] = pd.to_datetime(merged_df["ended_at"], format='%Y-%m-%d %H:%M:%S')
```

#### Finding the numerical and categorical columns. If the dtype is "O" then add it to the categorical_columns list else add it to the numerical_columns list

```python
categorical_columns = []
numerical_columns   = []
for col in merged_df.columns:
    if merged_df[col].dtype == 'O':
        categorical_columns.append(col)
    else:
        numerical_columns.append(col)
```

```python
numerical_columns
```

#### To determine the ride length, the calculation involves subtracting the "started_at" time from the "ended_at" time. 
###### It's important to note that the calculation includes extracting the time difference in seconds and dividing it by 60 to obtain the ride length in minutes. Additionally, the resulting value is rounded to two decimal places.

```python
merged_df['ride_length'] = round((merged_df['ended_at'] - merged_df['started_at']).dt.seconds/60, 2)
```

```python
merged_df.head()
```

```python
# The lambda function is used to format each statistic as a floating-point number.
statistics_ride_length = merged_df['ride_length'].describe().apply(lambda x: format(x, 'f'))
```

```python
statistics_ride_length
```

```python
ride_length_des_table = pd.DataFrame(statistics_ride_length)
```

#### Creating some more columns for month, day and hour

```python
merged_df['month'] = merged_df['started_at'].dt.month_name()
merged_df['day']   = merged_df['started_at'].dt.day_name()
merged_df['hour']  = merged_df['started_at'].dt.hour
```

```python
merged_df.head()
```

#### Converting the month and day columns to categorical data to sort the values properly on x-axis during the visualization

```python
month_names = [pd.to_datetime(x, format = '%m').month_name() for x in range(1, 13)]
merged_df['month'] = pd.Categorical(merged_df['month'], categories=month_names)

day_names = [pd.to_datetime(x, format = '%d').day_name() for x in range(1, 8)]
merged_df['day'] = pd.Categorical(merged_df['day'], categories=day_names)
```

## A) Based on the initial question we posed, we would like to understand the differences in how annual members and casual riders utilize Cyclistic bikes. To achieve this, we need to ascertain the number of casual riders and member riders.

```python
plot_color = 'Greens'
```

```python
plot_color2 = 'coolwarm'
```

```python
plt.figure(figsize=(20, 8), dpi=300)
plt.ticklabel_format(style='plain', axis='y')
data = merged_df.groupby(['member_casual']).size()
pal = sns.color_palette(plot_color, len(data))
rank = data.argsort().argsort()
ax = sns.barplot(x=data.index, y=data, palette=np.array(pal)[rank])
ax.set(xlabel=None)
[ax.bar_label(x, fmt='%d', fontsize=10) for x in ax.containers]
ticks = ax.set_xticklabels(ax.get_xticklabels(), fontsize=15)
```

* Based on the bar diagram above, it is evident that the number of member riders is significantly higher than that of casual riders.

## B) Determine in which month the member rides the most

```python
plt.figure(figsize=(20, 8), dpi=300)
data = merged_df.groupby('month').size()

# Define the desired order of months
month_order = ['May', 'June', 'July', 'August', 'September', 'October',
               'November', 'December', 'January', 'February', 'March', 'April']

# Reindex the data based on the desired month order
data = data.reindex(month_order)
pal = sns.color_palette(plot_color, len(data))
rank = data.argsort().argsort() 
ax = sns.barplot(x=data.index, y=data, palette=np.array(pal)[rank])
ax.set(xlabel=None)
[ax.bar_label(x, fontsize = 10) for x in ax.containers]
ticks = ax.set_xticklabels(ax.get_xticklabels(), fontsize = 15)
```

* June, July, and August were the busiest months during this annual period.

## C) Lets find out the busiest week day of this annual period

```python
plt.figure(figsize=(20,8), dpi = 300)
data = merged_df.groupby(['day']).size()
pal = sns.color_palette(plot_color, len(data))
rank = data.argsort().argsort()
ax = sns.barplot(x = data.index, y = data, palette=np.array(pal)[rank])
[ax.bar_label(x, fontsize = 10) for x in ax.containers]
ax.set(xlabel = None)
ticks = ax.set_xticklabels(ax.get_xticklabels())
ticks = ax.set_xticklabels(ax.get_xticklabels(), fontsize = 15)
```

## D) Lets find out the busiest hour of this annual period
